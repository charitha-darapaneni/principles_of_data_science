### Bootstrap

```python
num_samples = 500
observations = 150
```

```python
bootstrap_mean = np.zeros(num_samples)
bootstrap_std_dev = np.zeros(num_samples)
bootstrap_percentile = np.zeros(num_samples)
```

```python
for i in range(num_samples):
    bootstrap_sample = np.random.choice(data['BloodPressure'], size=observations, replace = True)
    bootstrap_sample_mean = np.mean(bootstrap_sample)
    bootstrap_sample_std = np.std(bootstrap_sample)
    bootstrap_sample_percentile = np.percentile(bootstrap_sample,98)
```

```python
population_mean = data['BloodPressure'].mean()
population_std = data['BloodPressure'].std()
population_percentile = np.percentile(data['BloodPressure'],98)
```

```python
labels=['bootstrap mean', 'population mean']
```

```python
values = [bootstrap_sample_mean, population_mean]
```

```python
mpl.bar(labels, values, color=['orange','green'])
mpl.title('Blood Pressure Mean comparison')
mpl.ylabel('Blood Pressure')
mpl.show()
```

```python
labels = ['bootstrap standard deviation', 'population standard deviation']
values = [bootstrap_sample_std, population_std]
```

```python
mpl.bar(labels, values, color=['orange','green'])
mpl.title('Blood Pressure Standard Deviation comparison')
mpl.ylabel('Blood Pressure')
mpl.show()
```

```python
labels = ['bootstrap percentile', 'population percentile']
values = [bootstrap_sample_percentile, population_percentile]
```

```python
mpl.bar(labels, values, color=['orange','green'])
mpl.title('Blood Pressure Percentile comparison')
mpl.ylabel('Blood Pressure')
mpl.show()
```
